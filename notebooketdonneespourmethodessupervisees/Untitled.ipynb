```python
auto
```

```python
auto['catPrix']=pd.qcut(auto['PRIX'],q=[0, 1/3, 2/3, 1], labels = ['Low','Medium','High'])
```

```python
auto
```

```python
y = auto[['catPrix']]
y
```

```python
autoFeatures=auto[['CYLINDRE','PUISSANCE','LONGUEUR','LARGEUR','POIDS','VITESSE']]
autoFeatures
```

```python
X=autoFeatures
```

```python
import numpy as np 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.10,random_state=0)
# print(X_train.shape,X_test.shape, y_train.shape, y_test.shape)
```

```python
print('X_train,sample')
print(X_train)
print('\n')
print('X_test,sample')
print(X_test)
print('\n')
print('y_train,sample')
print(y_train)
print('\n')
print('y_test,sample')
print(y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier 
from sklearn.tree.export import export_text 
clf=DecisionTreeClassifier()
clf=clf.fit(X_train,y_train)
r = export_text(clf, feature_names=list(autoFeatures.columns))
print(r)
```

```python
y_pred=clf.predict(X_test)
y_pred
```

```python
y_pred=y_pred.reshape(y_test.shape)
```

```python
misclassified = (y_test != y_pred.reshape(y_test.shape)).sum()
print('Misclassified samples: %d' % misclassified)
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy: %.2f' % accuracy_score(y_test, y_pred))
```
