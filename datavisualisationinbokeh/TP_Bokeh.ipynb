```python
from math import pi
import pandas as pd
from bokeh.palettes import Category20c
from bokeh.transform import cumsum
x = { 'United States': 157, 'United Kingdom': 93, 'Japan': 89, 'China': 63,
      'Germany': 44, 'India': 42, 'Italy': 40, 'Australia': 35, 'Brazil': 32,
      'France': 31, 'Taiwan': 31, 'Spain': 29 }
data = pd.Series(x).reset_index(name='value').rename(columns={'index':'country'})
data['color'] = Category20c[len(x)]
data['angle'] = data['value']/data['value'].sum() * 2*pi
p = figure(plot_height=350, title="Pie Chart", toolbar_location=None,
           tools="hover", tooltips="@country: @value")

p.wedge(x=0, y=1, radius=0.4, 
        
        # use cumsum to cumulatively sum the values for start and end angles
        start_angle=cumsum('angle', include_zero=True), end_angle=cumsum('angle'),
        line_color="white", fill_color='color', legend_field='country', source=data)

p.axis.axis_label=None
p.axis.visible=False
p.grid.grid_line_color = None

show(p)
```

```python
from bokeh.models.annotations import BoxAnnotation
```

```python
# Ce module BoxAnnotation permet de mettre des box sur la figure un peu partout sur la figure 
x=np.linspace(0,20,200)
y=np.sin(x)
p=figure(y_range=(-2,2))
p.line(x,y)
upper =BoxAnnotation(bottom=1,fill_alpha=0.1,fill_color='olive')
p.add_layout(upper)
lower=BoxAnnotation(top=-1,fill_alpha=0.1,fill_color='grey')
p.add_layout(lower)
center = BoxAnnotation(top=0.6, bottom=-0.3, left=7, right=12, fill_alpha=0.1, fill_color='navy')
p.add_layout(center)

#la fonction prend comme argument labscisse a partir de laquelle le box commence , la couleur du box et la taille
#et faut preciser la position du box ........ 


show(p)
```
