```python
def eigenvalues_de_Kaiser(egvalues,egvectors):
   liste=[]
   liste1=[]
   for i in range(len(egvalues)):
       if egvalues[i]>1:
           liste.append(egvalues[i])
           liste1.append(egvectors[i])
   return liste,liste1
```
